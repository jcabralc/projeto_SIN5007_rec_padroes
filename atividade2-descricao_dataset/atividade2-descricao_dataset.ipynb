## Desbalanceamento

```python
total = len(df)

fig, ax = plt.subplots(figsize=(15,6))
g = sns.countplot(data=df, x='Biopsy')
for p in g.patches:
    height = p.get_height()
    g.text(p.get_x()+p.get_width()/2.,
            height/2,
            '{:1.4f}'.format(100*height/total),
            ha="center", color='white') 
plt.show()
```

Dados completamento desbalanceados
