```python
# imports

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

## Import do Dataset

```python
df = pd.read_csv('../data/kag_risk_factors_cervical_cancer.csv')

print(df.shape)

df.head()
```

```python
# Tipos de dados de cada feature
df.dtypes
```

## Missing Values

```python
# Procurando por valores NA da forma habitual
#display(df.isnull().sum())

# verificando por ?
#df[df.isin(["?"]).any(axis=1)]

# substituindo ? por NAN
df.replace('?', np.NAN,inplace=True)

df.isnull().sum()
```

Features STDs: Time since first diagnosis e STDs: Time since last diagnosis possuem cerca de 91,7% de dados missing....

O restante das features possuem cerca de 12% de dados missing

## Desbalanceamento

```python
total = len(df)

fig, ax = plt.subplots(figsize=(15,6))
g = sns.countplot(data=df, x='Biopsy')
for p in g.patches:
    height = p.get_height()
    g.text(p.get_x()+p.get_width()/2.,
            height/2,
            '{:1.4f}'.format(100*height/total),
            ha="center", color='white') 
plt.show()
```

Dados completamento desbalanceados
